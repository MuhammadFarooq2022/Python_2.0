# 03- How to rename columns

```python
df = pd.DataFrame({"A col": [1,2,3,6,7,8,9,22], "B col": [4,5,6,7,8,9,10,11]})
df
```

```python
df.rename(columns={"A col": "col_a", "B col": "col_b"}, inplace=True)
df
```

```python
# rename columns
df.columns=["col_aa", "col_bb"]
df
```

```python
# To replace any character, string
df.columns = df.columns.str.replace("_", " ")
df
```

```python
df.columns = df.columns.str.replace(" ", "_")
df
```

```python
# Adding prefix to columns
df = df.add_prefix("baba_")
df
```

```python
# Adding suffix to columns
df = df.add_suffix("_baba")
df
```

# 16- Splitting a string into multiple columns

```python
import pandas as pd

df = pd.DataFrame({"name":["John Arthur Doe", "Jane Ann Smith"],
                    "location":["Los Angeles, CA", "Washington, DC"]}) 
df
```

```python
df.name.str.split(" ").head()
```

```python
# Splitting into three columns
df.name.str.split(" ",expand=True)
```

```python
# Adding those splits into new columns
df[["first_name", "middle_name", "last_name"]]=df.name.str.split(" ",expand=True)
```

```python
df
```

```python
df.location.str.split(", ",expand=True)
```

```python
df[["City", "State"]]=df.location.str.split(", ",expand=True)
```

```python
df
```

```python
# Refine data manipulation
df = df[["first_name", "middle_name", "last_name", "City", "State"]]
df
```
